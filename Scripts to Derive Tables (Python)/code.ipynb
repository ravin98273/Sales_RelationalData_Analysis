```python
import pandas as pd
import numpy as np
```

```python
try:
    df = pd.read_csv('..\data\sales_data_sample.csv', encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv('..\data\sales_data_sample.csv', encoding='ISO-8859-1')
    except UnicodeDecodeError:
        df = pd.read_csv('..\data\sales_data_sample.csv', encoding='cp1252')
```

```python
location = df[['ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY']].drop_duplicates()

location['LOCATION_ID'] = ['L' + str(i) for i in range(1, len(location) + 1)]
df_m = pd.merge(df, location, on=['ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY'], how='left')
```

```python
time = df[['ORDERDATE', 'STATUS', 'QTR_ID', 'MONTH_ID', 'YEAR_ID']].drop_duplicates()

time['TIME_ID'] = ['T' + str(i) for i in range(1, len(time) + 1)]
df_m = pd.merge(df_m, time, on=['ORDERDATE', 'STATUS', 'QTR_ID', 'MONTH_ID', 'YEAR_ID'], how='left')
```

```python
product = df[['PRODUCTLINE', 'MSRP', 'PRODUCTCODE']].drop_duplicates()

product['PRODUCT_ID'] = ['P' + str(i) for i in range(1, len(product) + 1)]
df_m = pd.merge(df_m, product, on=['PRODUCTLINE', 'MSRP', 'PRODUCTCODE'], how='left')
```

```python
customer = df[['CUSTOMERNAME', 'PHONE','CONTACTLASTNAME', 'CONTACTFIRSTNAME','ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY']].drop_duplicates()
customer['CUSTOMER_ID'] = ['C' + str(i) for i in range(1, len(customer) + 1)]
df_m = pd.merge(df_m, customer, on=['CUSTOMERNAME', 'PHONE','CONTACTLASTNAME', 'CONTACTFIRSTNAME','ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY'], how='left')

customer = pd.merge(customer, location, on=['ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY'], how='left')
```

```python
sales=df_m[['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER','SALES','LOCATION_ID', 'TIME_ID', 'PRODUCT_ID', 'CUSTOMER_ID']]
```
